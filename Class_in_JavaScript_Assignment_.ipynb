Problem Statement 1: Create a Constructor for a Person

Write a JavaScript function constructor named Person that takes two parameters, name, and age, and Add a
method to the prototype to display name and age. Then, create two instances of Person and display their
names and ages.

````python
Here’s a complete solution for **Task 1: Creating a `Person` constructor** with a prototype method in JavaScript:

---

## ✅ JavaScript Code

```javascript id="p1constructor"
function Person(name, age) {
  this.name = name;
  this.age = age;
}

// Add a method to the prototype to display name and age
Person.prototype.displayInfo = function() {
  console.log(`Name: ${this.name}, Age: ${this.age}`);
};

// Create two instances of Person
const person1 = new Person("Alice", 25);
const person2 = new Person("Bob", 30);

// Display their information
person1.displayInfo(); // Output: Name: Alice, Age: 25
person2.displayInfo(); // Output: Name: Bob, Age: 30
```

---

## 🔍 Explanation

1. **Constructor Function**

   * `Person(name, age)` is a function constructor.
   * `this.name` and `this.age` store instance-specific values.

2. **Prototype Method**

   * `Person.prototype.displayInfo` is added to the prototype so **all instances share the same function**, saving memory.

3. **Creating Instances**

   * `new Person("Alice", 25)` creates an object with the specified properties.

4. **Calling Method**

   * `person1.displayInfo()` prints the name and age to the console.

---

## 💡 Output

```id="p1output"
Name: Alice, Age: 25
Name: Bob, Age: 30
````

Problem Statement 2: Implement a Bank Account

Create a constructor function named BankAccount that initializes a bank account with an initial balance.
Include methods for depositing and withdrawing money from the account. Write code to demonstrate these
operations on a bank account object.

````python
Here’s a complete solution for **Task 2: Implementing a Bank Account** using a **constructor function** in JavaScript:

---

## ✅ JavaScript Code

```javascript id="bankaccount"
function BankAccount(initialBalance) {
  this.balance = initialBalance;
}

// Deposit method
BankAccount.prototype.deposit = function(amount) {
  if (amount > 0) {
    this.balance += amount;
    console.log(`Deposited: $${amount}. New Balance: $${this.balance}`);
  } else {
    console.log("Deposit amount must be positive!");
  }
};

// Withdraw method
BankAccount.prototype.withdraw = function(amount) {
  if (amount > 0 && amount <= this.balance) {
    this.balance -= amount;
    console.log(`Withdrew: $${amount}. New Balance: $${this.balance}`);
  } else if (amount > this.balance) {
    console.log("Insufficient balance!");
  } else {
    console.log("Withdrawal amount must be positive!");
  }
};

// Display current balance
BankAccount.prototype.displayBalance = function() {
  console.log(`Current Balance: $${this.balance}`);
};

// Create a bank account instance
const myAccount = new BankAccount(500);

// Demonstrate operations
myAccount.displayBalance();   // Current Balance: $500
myAccount.deposit(200);       // Deposited: $200. New Balance: $700
myAccount.withdraw(100);      // Withdrew: $100. New Balance: $600
myAccount.withdraw(700);      // Insufficient balance!
myAccount.displayBalance();   // Current Balance: $600
```

---

## 🔍 Explanation

1. **Constructor Function**

   * `BankAccount(initialBalance)` creates a new bank account with a starting balance.

2. **Deposit Method**

   * Adds money to `balance`
   * Checks that amount is positive

3. **Withdraw Method**

   * Deducts money from `balance`
   * Checks that amount is positive and ≤ current balance

4. **Display Method**

   * Shows current balance

5. **Prototype Usage**

   * Methods are on `BankAccount.prototype` so **all instances share them**, reducing memory usage.

---

## 💡 Example Output

```id="bankoutput"
Current Balance: $500
Deposited: $200. New Balance: $700
Withdrew: $100. New Balance: $600
Insufficient balance!
Current Balance: $600
````

Problem Statement 5: Implement a Calculator Class

Create a class called Calculator that initializes two values value1 and value2 to store numbers. Add methods for
add, subtract, multiply, and divide. Perform and show operations.

````python
Here’s a complete solution for **Task 5: Implementing a Calculator class** using ES6 syntax in JavaScript:

---

## ✅ JavaScript Code

```javascript id="calculatorclass"
class Calculator {
  // Constructor to initialize two numbers
  constructor(value1, value2) {
    this.value1 = value1;
    this.value2 = value2;
  }

  // Method to add
  add() {
    return this.value1 + this.value2;
  }

  // Method to subtract
  subtract() {
    return this.value1 - this.value2;
  }

  // Method to multiply
  multiply() {
    return this.value1 * this.value2;
  }

  // Method to divide
  divide() {
    if (this.value2 !== 0) {
      return this.value1 / this.value2;
    } else {
      return "Error: Division by zero!";
    }
  }
}

// Create an instance of Calculator
const calc = new Calculator(10, 5);

// Perform operations and display results
console.log(`Addition: ${calc.add()}`);       // 15
console.log(`Subtraction: ${calc.subtract()}`); // 5
console.log(`Multiplication: ${calc.multiply()}`); // 50
console.log(`Division: ${calc.divide()}`);     // 2
```

---

## 🔍 Explanation

1. **Class Definition**

   * `class Calculator` defines a blueprint for a calculator object.

2. **Constructor**

   * `constructor(value1, value2)` initializes two numbers for calculations.

3. **Methods**

   * `add()`, `subtract()`, `multiply()`, `divide()` perform arithmetic operations.
   * `divide()` includes a **check for division by zero**.

4. **Instance**

   * `new Calculator(10, 5)` creates a calculator object with numbers 10 and 5.

5. **Output**

   * Calling methods returns results of operations.

---

## 💡 Output

```id="calcoutput"
Addition: 15
Subtraction: 5
Multiplication: 50
Division: 2
````

Problem Statement 6: Design a Class for a Geometric Shape

Design a class called Shape that can represent various geometric shapes. Implement subclasses for specific
shapes like Circle, Rectangle, and RightTriangle. Each subclass should have properties of that shape (e.g..
Width, height for a rectangle) and methods for calculating the area and perimeter of the shape. Create
instances of these shapes and calculate their areas and perimeters.




````python
Here’s a complete solution for **Task 6: Designing a Shape class with subclasses** in JavaScript using ES6 classes:

---

## ✅ JavaScript Code

```javascript id="shapeclass"
class Shape {
  constructor(name) {
    this.name = name;
  }

  // Placeholder methods for area and perimeter
  area() {
    return 0;
  }

  perimeter() {
    return 0;
  }
}

// Circle subclass
class Circle extends Shape {
  constructor(radius) {
    super("Circle");
    this.radius = radius;
  }

  area() {
    return Math.PI * this.radius ** 2;
  }

  perimeter() {
    return 2 * Math.PI * this.radius;
  }
}

// Rectangle subclass
class Rectangle extends Shape {
  constructor(width, height) {
    super("Rectangle");
    this.width = width;
    this.height = height;
  }

  area() {
    return this.width * this.height;
  }

  perimeter() {
    return 2 * (this.width + this.height);
  }
}

// RightTriangle subclass
class RightTriangle extends Shape {
  constructor(base, height) {
    super("RightTriangle");
    this.base = base;
    this.height = height;
  }

  area() {
    return 0.5 * this.base * this.height;
  }

  perimeter() {
    const hypotenuse = Math.sqrt(this.base ** 2 + this.height ** 2);
    return this.base + this.height + hypotenuse;
  }
}

// Create instances and calculate areas and perimeters
const circle = new Circle(5);
console.log(`Circle - Area: ${circle.area().toFixed(2)}, Perimeter: ${circle.perimeter().toFixed(2)}`);

const rectangle = new Rectangle(4, 6);
console.log(`Rectangle - Area: ${rectangle.area()}, Perimeter: ${rectangle.perimeter()}`);

const triangle = new RightTriangle(3, 4);
console.log(`RightTriangle - Area: ${triangle.area()}, Perimeter: ${triangle.perimeter().toFixed(2)}`);
```

---

## 🔍 Explanation

1. **Base Class: Shape**

   * Provides a common structure (`name`) for all shapes.
   * Placeholder methods `area()` and `perimeter()` can be overridden.

2. **Circle Subclass**

   * Property: `radius`
   * `area()` → π × r²
   * `perimeter()` → 2 × π × r

3. **Rectangle Subclass**

   * Properties: `width`, `height`
   * `area()` → width × height
   * `perimeter()` → 2 × (width + height)

4. **RightTriangle Subclass**

   * Properties: `base`, `height`
   * `area()` → 0.5 × base × height
   * `perimeter()` → base + height + hypotenuse

5. **Creating Instances**

   * `circle`, `rectangle`, `triangle` are created with their dimensions.
   * Methods calculate areas and perimeters.

---

## 💡 Sample Output

```id="shapeoutput"
Circle - Area: 78.54, Perimeter: 31.42
Rectangle - Area: 24, Perimeter: 20
RightTriangle - Area: 6, Perimeter: 12.00
````

Problem Statement 7: Implement a Library System

Create a class called Library that can manage a collection of books. Each book should be represented as an
instance of a Book class (with properties like title, author, and availability). Implement methods in the Library
class to add books, check out books, and return books. Keep track of the books available.

````python
Here’s a complete solution for **Task 7: Implementing a Library System** using JavaScript ES6 classes:

---

## ✅ JavaScript Code

```javascript id="librarysystem"
class Book {
  constructor(title, author) {
    this.title = title;
    this.author = author;
    this.isAvailable = true; // Book availability
  }

  // Display book info
  displayInfo() {
    console.log(`Title: ${this.title}, Author: ${this.author}, Available: ${this.isAvailable}`);
  }
}

class Library {
  constructor() {
    this.books = []; // Collection of Book objects
  }

  // Add a book to the library
  addBook(book) {
    this.books.push(book);
    console.log(`Added "${book.title}" to the library.`);
  }

  // Check out a book
  checkoutBook(title) {
    const book = this.books.find(b => b.title === title);
    if (book && book.isAvailable) {
      book.isAvailable = false;
      console.log(`You have checked out "${book.title}".`);
    } else if (book) {
      console.log(`Sorry, "${book.title}" is currently not available.`);
    } else {
      console.log(`Book "${title}" not found in the library.`);
    }
  }

  // Return a book
  returnBook(title) {
    const book = this.books.find(b => b.title === title);
    if (book) {
      book.isAvailable = true;
      console.log(`You have returned "${book.title}".`);
    } else {
      console.log(`Book "${title}" not found in the library.`);
    }
  }

  // Display all books
  displayBooks() {
    console.log("Library Collection:");
    this.books.forEach(book => book.displayInfo());
  }
}

// Example Usage

// Create a library
const myLibrary = new Library();

// Create books
const book1 = new Book("To Kill a Mockingbird", "Harper Lee");
const book2 = new Book("1984", "George Orwell");
const book3 = new Book("The Great Gatsby", "F. Scott Fitzgerald");

// Add books to the library
myLibrary.addBook(book1);
myLibrary.addBook(book2);
myLibrary.addBook(book3);

// Display all books
myLibrary.displayBooks();

// Checkout a book
myLibrary.checkoutBook("1984");

// Attempt to checkout the same book again
myLibrary.checkoutBook("1984");

// Return a book
myLibrary.returnBook("1984");

// Display updated books
myLibrary.displayBooks();
```

---

## 🔍 Explanation

1. **Book Class**

   * Properties: `title`, `author`, `isAvailable`
   * Method: `displayInfo()` to show book details

2. **Library Class**

   * Property: `books` (array of Book objects)
   * Methods:

     * `addBook(book)` → adds a new book
     * `checkoutBook(title)` → marks book as unavailable
     * `returnBook(title)` → marks book as available
     * `displayBooks()` → lists all books with their availability

3. **Example Workflow**

   * Add books to the library
   * Checkout and return books
   * Display the library collection before and after operations

---

## 💡 Sample Output

```id="libraryoutput"
Added "To Kill a Mockingbird" to the library.
Added "1984" to the library.
Added "The Great Gatsby" to the library.
Library Collection:
Title: To Kill a Mockingbird, Author: Harper Lee, Available: true
Title: 1984, Author: George Orwell, Available: true
Title: The Great Gatsby, Author: F. Scott Fitzgerald, Available: true
You have checked out "1984".
Sorry, "1984" is currently not available.
You have returned "1984".
Library Collection:
Title: To Kill a Mockingbird, Author: Harper Lee, Available: true
Title: 1984, Author: George Orwell, Available: true
Title: The Great Gatsby, Author: F. Scott Fitzgerald, Available: true
````
