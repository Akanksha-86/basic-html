1. explain the basic structure of HTML page. why is it important to follow this structure?
     - The basic structure of an HTML page ensures proper rendering and interpretation by web browsers. it's built upon key elements like the **DOCTYPE **  declaration, <html>, <head> , and <body> tages, The **DOCTYPE ** specifies the HTML version, the <head> contains metadata (like title and links to external resources), and the <body> holds the visible content.

*  ** importance of the structure**
*   consistency: The <body> contains the actual content of the page, including text, images, and other interactive elements, which are displayed to the user.


*   Browser compatibility: Adhnering to the basic HTML structure ensures that page will render consistently accross differently.


*   Accessibilty: A well strutured page is easier for assistive technologies, like screen reader, to interpret and provide content to user with disabilities.


*   Semantic meaning: HTML structure allows for sementic meaning to be conveyed, helping search engines understand the content of the page better.
