Q4. Using any of the Pseudo element selectors, add any random words after and before the given HTML tag
Assignment Question
*  An example should be one HTML button tac
*  When hovering the color should be changed and the shadow of the button should be added^
*  Proper padding space should be given^
*  The cursor style should also be changed to a pointer on hover

image.png


```python
/* html file */
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Pseudo Elements Button</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <button class="magic-btn">Click Me!</button>
</body>
</html>

/* css file */
/* Pseudo-element selectors for button styling */

.magic-btn {
    position: relative;
    background: linear-gradient(45deg, #ff6b6b, #4ecdc4);
    color: white;
    border: none;
    padding: 15px 30px;
    font-size: 18px;
    font-weight: bold;
    border-radius: 50px;
    cursor: default;
    overflow: hidden;
    transition: all 0.3s ease;
    box-shadow: 0 4px 15px rgba(0,0,0,0.2);
}

/* BEFORE Pseudo-element: Add random words BEFORE button text */
.magic-btn::before {
    content: "✨ START ";
    position: absolute;
    left: 10px;
    top: 50%;
    transform: translateY(-50%);
    font-size: 14px;
    opacity: 0;
    transition: all 0.3s ease;
}

/* AFTER Pseudo-element: Add random words AFTER button text */
.magic-btn::after {
    content: " END ✨";
    position: absolute;
    right: 10px;
    top: 50%;
    transform: translateY(-50%);
    font-size: 14px;
    opacity: 0;
    transition: all 0.3s ease;
}

/* HOVER Effects */
.magic-btn:hover {
    background: linear-gradient(45deg, #4ecdc4, #ff6b6b);
    color: white;
    transform: translateY(-3px);
    box-shadow: 0 8px 25px rgba(0,0,0,0.4);
    cursor: pointer;
}

/* Show pseudo-elements on hover */
.magic-btn:hover::before,
.magic-btn:hover::after {
    opacity: 1;
    transform: translateY(-50%) scale(1.05);
}

/* Ripple effect using pseudo-element */
.magic-btn::before {
    z-index: 0;
}

.magic-btn::after {
    z-index: 0;
}

.magic-btn:active::before {
    content: "";
    position: absolute;
    width: 300px;
    height: 300px;
    background: rgba(255,255,255,0.3);
    border-radius: 50%;
    top: 50%;
    left: 50%;
    transform: translate(-50%, -50%) scale(0);
    animation: ripple 0.6s linear;
    opacity: 1;
}

@keyframes ripple {
    to {
        transform: translate(-50%, -50%) scale(4);
        opacity: 0;
    }
}
```
