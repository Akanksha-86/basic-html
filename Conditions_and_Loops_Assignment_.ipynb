# Conditions and Loops

Q1. What are conditional statements? Explain conditional statements with syntax and examples

 - **Conditional statements** are used in programming to **make decisions based on certain conditions**. They allow your program to execute different blocks of code depending on whether a condition is **true or false**.

---

## 🧠 Why They Are Used

Conditional statements help:

* Control program flow
* Perform different actions based on input
* Add logic and decision-making to applications

---

## 🔑 Common Types of Conditional Statements

---

### 1. `if` Statement

Executes code **only if the condition is true**

#### ✅ Syntax:

```javascript
if (condition) {
  // code to execute
}
```

#### ✅ Example:

```javascript
let age = 18;

if (age >= 18) {
  console.log("You can vote");
}
```

---

### 2. `if...else` Statement

Executes one block if true, another if false

#### ✅ Syntax:

```javascript
if (condition) {
  // true block
} else {
  // false block
}
```

#### ✅ Example:

```javascript
let age = 16;

if (age >= 18) {
  console.log("Adult");
} else {
  console.log("Minor");
}
```

---

### 3. `if...else if...else` Ladder

Used when you have **multiple conditions**

#### ✅ Syntax:

```javascript
if (condition1) {
  // code
} else if (condition2) {
  // code
} else {
  // default code
}
```

#### ✅ Example:

```javascript
let marks = 75;

if (marks >= 90) {
  console.log("Grade A");
} else if (marks >= 60) {
  console.log("Grade B");
} else {
  console.log("Grade C");
}
```

---

### 4. `switch` Statement

Used when checking **multiple fixed values**

#### ✅ Syntax:

```javascript
switch (expression) {
  case value1:
    // code
    break;
  case value2:
    // code
    break;
  default:
    // code
}
```

#### ✅ Example:

```javascript
let day = 2;

switch (day) {
  case 1:
    console.log("Monday");
    break;
  case 2:
    console.log("Tuesday");
    break;
  default:
    console.log("Other day");
}
```

---

## 🧠 Simple Analogy

Conditional statements are like making decisions in real life:

👉 “If it rains → take an umbrella
Else → go normally”

---

## 📌 In One Line

Conditional statements allow a program to **make decisions and execute code based on conditions**.


Q2. Write a program that grades students based on their marks^
8 If greater than 90 then A GradE
8 If between 70 and 90 then a B gradE
8 If between 50 and 70 then a C gradE
8 Below 50 then an F grade



````python
<!DOCTYPE html>
<html>
<head>
    <title>Student Grading System</title>
    <style>
        body { font-family: Arial, sans-serif; max-width: 500px; margin: 50px auto; padding: 20px; }
        input, button { padding: 12px; font-size: 16px; margin: 10px 0; }
        .result { background: #f0f8ff; padding: 20px; border-radius: 10px; margin-top: 20px; font-weight: bold; }
    </style>
</head>
<body>
    <h1>🎓 Student Grading System</h1>
    <input type="number" id="marks" placeholder="Enter marks (0-100)" min="0" max="100">
    <button onclick="calculateGrade()">Get Grade</button>

    <div id="result" class="result" style="display:none;"></div>

    <script>
    function calculateGrade() {
        const marks = parseFloat(document.getElementById('marks').value);

        if (isNaN(marks) || marks < 0 || marks > 100) {
            showResult("❌ Please enter valid marks (0-100)", "error");
            return;
        }

        let grade;
        if (marks > 90) {
            grade = "A 🎉 Excellent!";
        } else if (marks >= 70) {
            grade = "B 👍 Very Good!";
        } else if (marks >= 50) {
            grade = "C 👌 Good!";
        } else {
            grade = "F 😞 Needs Improvement";
        }

        showResult(`Marks: ${marks}/100<br>Grade: ${grade}`);
    }

    function showResult(message, type = "success") {
        const resultDiv = document.getElementById('result');
        resultDiv.innerHTML = message;
        resultDiv.className = `result ${type}`;
        resultDiv.style.display = 'block';
    }
    </script>
</body>
</html>
````

Q3. What are loops, and what do we need them? Explain different types of loops with their syntax and
examples
  
  - **Loops** are programming constructs that allow you to **repeat a block of code multiple times** until a specific condition is met.

---

## 🧠 Why Do We Need Loops?

Without loops, you would have to write the same code again and again. Loops help to:

* Save time and reduce code repetition
* Automate repetitive tasks
* Make code shorter and more efficient

👉 Example: Printing numbers from 1 to 100 would be very long without loops.

---

## 🔁 Types of Loops (JavaScript)

---

### 1. `for` Loop

Used when you know **how many times** you want to run the loop.

#### ✅ Syntax:

```javascript id="h0vd8y"
for (initialization; condition; update) {
  // code
}
```

#### ✅ Example:

```javascript id="z2e2nt"
for (let i = 1; i <= 5; i++) {
  console.log(i);
}
```

👉 Output: 1 2 3 4 5

---

### 2. `while` Loop

Used when the number of iterations is **not fixed** and depends on a condition.

#### ✅ Syntax:

```javascript id="c5v2gq"
while (condition) {
  // code
}
```

#### ✅ Example:

```javascript id="5j1m6p"
let i = 1;

while (i <= 5) {
  console.log(i);
  i++;
}
```

---

### 3. `do...while` Loop

Executes the code **at least once**, even if the condition is false.

#### ✅ Syntax:

```javascript id="y8wgr9"
do {
  // code
} while (condition);
```

#### ✅ Example:

```javascript id="gfbpsx"
let i = 1;

do {
  console.log(i);
  i++;
} while (i <= 5);
```

---

### 4. `for...of` Loop

Used to loop through **arrays or iterable objects**.

#### ✅ Syntax:

```javascript id="w0rmzg"
for (let item of iterable) {
  // code
}
```

#### ✅ Example:

```javascript id="yvr8m7"
let fruits = ["apple", "banana", "mango"];

for (let fruit of fruits) {
  console.log(fruit);
}
```

---

### 5. `for...in` Loop

Used to loop through **object properties (keys)**.

#### ✅ Syntax:

```javascript id="w3g9d3"
for (let key in object) {
  // code
}
```

#### ✅ Example:

```javascript id="l7c9qe"
let student = { name: "John", age: 20 };

for (let key in student) {
  console.log(key, student[key]);
}
```

---

## 🧠 Simple Analogy

Loops are like telling someone:

👉 “Repeat this task until I say stop.”

---

## 📌 In One Line

Loops are used to **execute a block of code repeatedly based on a condition**, making programs efficient and concise.

Q5. Use the while loop to print numbers from 1 to 25 in ascending and descending order.

````python
<!DOCTYPE html>
<html>
<head>
    <title>While Loop Numbers</title>
    <style>
        body { font-family: 'Courier New', monospace; padding: 40px; background: #1a1a2e; color: #00ff88; }
        pre { background: #16213e; padding: 20px; border-radius: 10px; margin: 20px 0; font-size: 18px; }
        button { padding: 12px 24px; font-size: 16px; background: #00ff88; color: #1a1a2e; border: none; border-radius: 8px; cursor: pointer; }
    </style>
</head>
<body>
    <h1>🔢 While Loop: 1 to 25</h1>
    <button onclick="printAscending()">Ascending (1 → 25)</button>
    <button onclick="printDescending()">Descending (25 → 1)</button>

    <pre id="output"></pre>

    <script>
    function printAscending() {
        let i = 1;
        let result = '';
        while (i <= 25) {
            result += i + (i < 25 ? ', ' : '');
            i++;
        }
        document.getElementById('output').textContent = result;
        console.log('Ascending:', result);
    }

    function printDescending() {
        let i = 25;
        let result = '';
        while (i >= 1) {
            result += i + (i > 1 ? ', ' : '');
            i--;
        }
        document.getElementById('output').textContent = result;
        console.log('Descending:', result);
    }
    </script>
</body>
</html>
````
