Problem Statement

Explain the purpose of using the var() function in css. Also you have created two buttons with id named
primaryBtn and secondaryBtn which should be given background colors using the var() function. The color
code for primaryColor is #00b7ff and secondaryColor is #6c757d.

```python
/* Html code */
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Variables Demo</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <div class="container">
        <button id="primaryBtn" class="btn">Primary Button</button>
        <button id="secondaryBtn" class="btn">Secondary Button</button>
    </div>
</body>
</html>

/* css code */

:root {
    --primaryColor: #00b7ff;
    --secondaryColor: #6c757d;
    --border-radius: 8px;
    --shadow: 0 4px 15px rgba(0,0,0,0.2);
    --transition: all 0.3s ease;
}


.btn {
    padding: 12px 30px;
    font-size: 16px;
    font-weight: 600;
    border: none;
    border-radius: var(--border-radius);
    cursor: pointer;
    transition: var(--transition);
    box-shadow: var(--shadow);
    margin: 0 10px;
}


#primaryBtn {
    background-color: var(--primaryColor);
    color: white;
}

#primaryBtn:hover {
    background-color: color-mix(in srgb, var(--primaryColor) 80%, #000);
    transform: translateY(-2px);
    box-shadow: 0 6px 20px rgba(0,183,255,0.4);
}

#secondaryBtn {
    background-color: var(--secondaryColor);
    color: white;
}

#secondaryBtn:hover {
    background-color: color-mix(in srgb, var(--secondaryColor) 80%, #000);
    transform: translateY(-2px);
    box-shadow: 0 6px 20px rgba(108,117,125,0.4);
}

.container {
    display: flex;
    justify-content: center;
    align-items: center;
    min-height: 100vh;
    background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
    gap: 20px;
    padding: 20px;
}

body {
    margin: 0;
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
}
```

Problem Statement

Create a 3D cube using the transform property of CSS.


image.png

```python
/* Html code */
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>3D CSS Cube</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <div class="scene">
        <div class="cube">
            <div class="face front">FRONT</div>
            <div class="face back">BACK</div>
            <div class="face right">RIGHT</div>
            <div class="face left">LEFT</div>
            <div class="face top">TOP</div>
            <div class="face bottom">BOTTOM</div>
        </div>
    </div>
</body>
</html>

/* css code */


* {
    margin: 0;
    padding: 0;
    box-sizing: border-box;
}

body {
    height: 100vh;
    background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
    display: flex;
    justify-content: center;
    align-items: center;
    font-family: 'Arial', sans-serif;
    perspective: 1000px; /* 3D Perspective */
    overflow: hidden;
}

.scene {
    width: 200px;
    height: 200px;
    perspective: 1000px; /* Creates 3D space */
    perspective-origin: 50% 50%;
}

.cube {
    width: 200px;
    height: 200px;
    position: relative;
    transform-style: preserve-3d; /* Essential for 3D children */
    transition: transform 0.5s ease;
    animation: rotateCube 10s infinite linear;
}

/* HOVER INTERACTION */
.cube:hover {
    animation-play-state: paused;
    transform: rotateX(20deg) rotateY(180deg) rotateZ(10deg);
}


.face {
    position: absolute;
    width: 200px;
    height: 200px;
    display: flex;
    align-items: center;
    justify-content: center;
    font-size: 24px;
    font-weight: bold;
    color: white;
    text-shadow: 2px 2px 4px rgba(0,0,0,0.5);
    border: 2px solid rgba(255,255,255,0.3);
}


.front  {
    background: linear-gradient(45deg, #ff6b6b, #ee5a52);
    transform: translateZ(100px);
}

.back   {
    background: linear-gradient(45deg, #4ecdc4, #44a08d);
    transform: rotateY(180deg) translateZ(100px);
}

.right  {
    background: linear-gradient(45deg, #feca57, #ff9ff3);
    transform: rotateY(90deg) translateZ(100px);
}

.left   {
    background: linear-gradient(45deg, #48dbfb, #0abde3);
    transform: rotateY(-90deg) translateZ(100px);
}

.top    {
    background: linear-gradient(45deg, #a8edea, #fed6e3);
    transform: rotateX(90deg) translateZ(100px);
}

.bottom {
    background: linear-gradient(45deg, #ffecd2, #fcb69f);
    transform: rotateX(-90deg) translateZ(100px);
}


@keyframes rotateCube {
    0% { transform: rotateX(0deg) rotateY(0deg) rotateZ(0deg); }
    100% { transform: rotateX(360deg) rotateY(360deg) rotateZ(360deg); }
}


@media (max-width: 768px) {
    .cube, .face {
        width: 150px;
        height: 150px;
    }

    .front, .back, .right, .left, .top, .bottom {
        transform: translateZ(75px);
    }

    .face {
        font-size: 18px;
    }
}
```

Task 3:


Problem Statement

Create a simple circular loader which will rotate continuously to look like a loading screen on a website.

image.png

As in the image we can see that the blue color is rotating continuously.

```python
/* Html code */

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Blue Circular Loader</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <div class="loader-container">
        <div class="loader"></div>
        <p>Loading...</p>
    </div>
</body>
</html>

/* css code */



* {
    margin: 0;
    padding: 0;
    box-sizing: border-box;
}

body {
    font-family: 'Arial', sans-serif;
    background: linear-gradient(135deg, #f5f3f3 0%, #f8f8fa 100%);
    height: 100vh;
    display: flex;
    justify-content: center;
    align-items: center;
    overflow: hidden;
}

.loader-container {
    display: flex;
    flex-direction: column;
    align-items: center;
    gap: 20px;
}

.loader {
    width: 80px;
    height: 80px;
    border: 8px solid white (226, 231, 233, 0.2); /* Transparent blue border */
    border-top: 8px solid #00b7ff; /* Solid blue top */
    border-radius: 50%;
    animation: spin 1s cubic-bezier(0.4, 0.0, 0.2, 1) infinite; /* Smooth rotation */
    box-shadow:
        0 0 20px blue (0, 183, 255, 0.4),
        inset 0 0 20px white(0, 183, 255, 0.1);
}

.loader::before {
    content: '';
    position: absolute;
    width: 20px;
    height: 20px;
    background: white ;
    border-radius: 50%;
    top: 50%;
    left: 50%;
    transform: translate(-50%, -50%);
    animation: pulse 1.5s ease-in-out infinite;
}

p {
    color: #00b7ff;
    font-size: 18px;
    font-weight: 500;
    text-shadow: 0 0 10px rgba(0, 183, 255, 0.5);
    animation: fadeInOut 2s ease-in-out infinite;
}

/* CORE ROTATION ANIMATION */
@keyframes spin {
    0% { transform: rotate(0deg); }
    100% { transform: rotate(360deg); }
}

/* PULSE EFFECT FOR CENTER DOT */
@keyframes pulse {
    0%, 100% {
        transform: translate(-50%, -50%) scale(1);
        opacity: 1;
    }
    50% {
        transform: translate(-50%, -50%) scale(1.2);
        opacity: 0.7;
    }
}

/* TEXT FADE ANIMATION */
@keyframes fadeInOut {
    0%, 100% { opacity: 0.6; }
    50% { opacity: 1; }
}

/* RESPONSIVE */
@media (max-width: 480px) {
    .loader {
        width: 60px;
        height: 60px;
        border-width: 6px;
    }
}
```

Task 5:


Problem Statement

You have to visit the PW Skills website (https://pwskills.com/) and have to change the content of the ‘Login /
Register’ button to ‘Connect with us’ using the developer tool. This should be done using the elements of
developer tools only, use of javascript is prohibited.

image.png

image.png

```python
/* css code for replace the login/register into conect with us */
/* Hide ALL Login/Register text */
*[class*="login"],
*[class*="Login"],
button:contains("Login"),
button:contains("Register"),
a:contains("Login") {
    font-size: 0 !important;
}

/* Replace with "Connect with us" */
*[class*="login"]::before,
button:contains("Login")::before,
a:contains("Login")::before {
    content: "Connect with us" !important;
    font-size: 14px !important;
    color: #fff !important;
}
```
