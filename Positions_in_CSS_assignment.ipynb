1. **problem statement**

write code the position 5 equal-sized(50 X 50px) boxes A, B,C,D, and E as follows,

1 box A 200px from the left, and 200px from the top of the viewport.
2 box B - 30px left and -30px above from the center of box A.
3  box B - 30px right and -30px above from the center of box A.
4  box B - 30px left and -30px below from the center of box A.
5  box B - 30px right and -30px below from the center of box A.

**Expected output**

image.png

```python
<!DOCTYPE html>
<html>
<head>
<style>
  body { margin: 0; padding: 0;
    border: solid black;
}
.square {
    border: solid black;
}
  .box {
    width: 50px;
    height: 50px;
    background-color: rgb(106, 241, 223);
    position: absolute;
    display: flex;
    justify-content: center;
    align-items: center;
    border: 1px solid rgb(106, 241, 223);
    color: black;
    font-weight: bold;
  }
  #boxA {
    z-index: 1;
    left: 200px;
    top: 200px;
  }
  /* Calculate center of Box A: (200+25, 200+25) = (225, 225) */
  /* For other boxes, their top/left should be relative to their own top-left corner */
  /* For example, if Box B's center is (225-30, 225-30) = (195, 195) */
  /* Then Box B's top-left corner is (195-25, 195-25) = (170, 170) */

  #boxB {
    left: calc(200px + 25px - 30px - 25px); /* center_x_A - 30 - half_width_B */
    top: calc(200px + 25px - 30px - 25px); /* center_y_A - 30 - half_height_B */
  }
  #boxC {
    left: calc(200px + 25px + 30px - 25px);
    top: calc(200px + 25px - 30px - 25px);
  }
  #boxD {
    left: calc(200px + 25px - 30px - 25px);
    top: calc(200px + 25px + 30px - 25px);
  }
  #boxE {
    left: calc(200px + 25px + 30px - 25px);
    top: calc(200px + 25px + 30px - 25px);
  }
</style>
</head>
<body>
    <div class="square">
<div id="boxA" class="box">A</div>
<div id="boxB" class="box">B</div>
<div id="boxC" class="box">C</div>
<div id="boxD" class="box">D</div>
<div id="boxE" class="box">E</div>
    </div>
</body>
</html>
```

5. explain the z-index, with a code explain.
- The CSS z-index property is used to control the stacking order of the positioned elements.

```python
<!DOCTYPE html>
<html lang="en">

    <head>
        <meta charset="UTF-8" />
        <meta name="viewport" content="width=device-width, initial-scale=1.0" />
        <link rel="stylesheet" href="style.css" />
        <title>CSS z-index</title>
    </head>

    <body>
        <div class="greenyellow">position: absolute; <br />z-index: 3;</div>
        <div class="orange">position: absolute; <br />z-index: 10;</div>
        <div class="skyblue">position: absolute; <br />z-index: 6;</div>
    </body>

</html>


div {
    font-size: 24px;
    padding: 12px;
}

div.greenyellow {
    position: absolute;
    top: 0;
    left: 0;

    /* specifying z-index value */
    z-index: 3;

    width: 300px;
    height: 280px;
    background-color: greenyellow;
}

div.orange {
    position: absolute;
    top: 190px;
    left: 0;

    /* specifying z-index value */
    z-index: 5;

    width: 220px;
    height: 150px;
    margin-top: -120px;
    background-color: orange;
}

div.skyblue {
    position: absolute;
    top: 120px;
    left: 160px;

    /* specifying z-index value */

    z-index: 6;
    width: 280px;
    height: 150px;
    padding-left: 120px;
    background-color: skyblue;
}
```
