```python
<!DOCTYPE html>
<html>
<head>
    <title>JavaScript Single-Threaded Demo</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 40px; }
        .task { padding: 10px; margin: 10px 0; border-radius: 5px; }
        .blocking { background: #ff6b6b; color: white; }
        .waiting { background: #feca57; color: white; }
        .completed { background: #51cf66; color: white; }
        #output { background: #f8f9fa; padding: 20px; border-radius: 8px; font-family: monospace; }
        button { padding: 10px 20px; font-size: 16px; margin: 10px; cursor: pointer; }
    </style>
</head>
<body>
    <h1>🧵 JavaScript Single-Threaded Nature Demo</h1>

    <div>
        <button onclick="startDemo()">🚀 Start Blocking Demo</button>
        <button onclick="clearOutput()">🗑️ Clear Output</button>
    </div>

    <div class="task blocking">
        🛑 BLOCKING TASK: 5-second synchronous loop (blocks Event Loop)
    </div>
    <div class="task waiting">
        ⏳ WAITING TASKS: setTimeout, setInterval, DOM events (queued)
    </div>

    <div id="output"></div>

    <script>
        const output = document.getElementById('output');

        function log(message) {
            const timestamp = new Date().toLocaleTimeString();
            output.innerHTML += `<div>[${timestamp}] ${message}</div>`;
            output.scrollTop = output.scrollHeight;
            console.log(`[${timestamp}] ${message}`);
        }

        function startDemo() {
            log('🎬 Demo started!');
            log('📋 Task queue before blocking:');

            // Task 1: Scheduled IMMEDIATELY but WAITS (setTimeout 0ms)
            setTimeout(() => {
                log('✅ setTimeout(0ms) FINALLY executes!');
            }, 0);

            // Task 2: Scheduled for 1 second later
            setTimeout(() => {
                log('✅ setTimeout(1000ms) executes!');
            }, 1000);

            // Task 3: Interval every 500ms (will queue up)
            const interval = setInterval(() => {
                log('🔄 setInterval tick');
            }, 500);

            // Task 4: Button click simulation
            document.querySelector('button').onclick = () => {
                log('🖱️ Button clicked during blocking!');
            };

            log('⏰ All tasks scheduled. NOW BLOCKING...');

            // 🔥 BLOCKING SYNCHRONOUS TASK (5 seconds)
            log('🚫 BLOCKING LOOP STARTS (5 seconds of CPU intensive work)');
            const startTime = Date.now();
            let counter = 0;

            // This SYNCHRONOUS loop BLOCKS the entire thread!
            while (Date.now() - startTime < 5000) {
                counter++;
                // Simulates heavy computation
            }

            log(`🚫 BLOCKING LOOP ENDS (counter: ${counter.toLocaleString()})`);
            clearInterval(interval);

            // Only AFTER blocking ends do queued tasks execute
            log('🎉 Event Loop UNBLOCKED! All queued tasks now execute.');
            log('🏁 Demo complete!');
        }

        function clearOutput() {
            output.innerHTML = '';
        }
    </script>
</body>
</html>

    /*console output during demo*/
                   [12:34:56] 🎬 Demo started!
[12:34:56] 📋 Task queue before blocking:
[12:34:56] ⏰ All tasks scheduled. NOW BLOCKING...
[12:34:56] 🚫 BLOCKING LOOP STARTS (5 seconds of CPU intensive work)
[12:35:01] 🚫 BLOCKING LOOP ENDS (counter: 12,345,678)
[12:35:01] ✅ setTimeout(0ms) FINALLY executes!  ← WAITED 5+ seconds!
[12:35:01] 🔄 setInterval tick                    ← Multiple ticks queued
[12:35:01] 🔄 setInterval tick
[12:35:01] ✅ setTimeout(1000ms) executes!     ← WAITED 5+ seconds!
[12:35:01] 🎉 Event Loop UNBLOCKED!
```

3.Chaining Promises with setTimeout

 Modify the delay function to chain multiple promises so that three messages are logged in sequence with

```python
<!DOCTYPE html>
<html>
<head>
    <title>Promise Chaining with setTimeout</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 40px; line-height: 1.6; }
        #output {
            background: #f8f9fa;
            padding: 20px;
            border-radius: 8px;
            min-height: 200px;
            font-family: monospace;
            border-left: 4px solid #007bff;
        }
        button { padding: 12px 24px; font-size: 16px; cursor: pointer; }
        .message { margin: 10px 0; padding: 8px; border-radius: 4px; }
        .msg1 { background: #d4edda; color: #155724; }
        .msg2 { background: #fff3cd; color: #856404; }
        .msg3 { background: #f8d7da; color: #721c24; }
    </style>
</head>
<body>
    <h1>🔗 Promise Chaining with setTimeout</h1>
    <p>Three messages logged in sequence: 1s → 2s → 3s delays</p>

    <button onclick="chainPromises()">🚀 Chain 3 Promises</button>
    <button onclick="clearOutput()">🗑️ Clear</button>

    <div id="output"></div>

    <script>
        const output = document.getElementById('output');

        // 🛠️ Delay function returns a Promise
        function delay(ms, message) {
            return new Promise((resolve) => {
                setTimeout(() => {
                    log(message);
                    resolve(message); // ✅ Resolve with message
                }, ms);
            });
        }

        // 🔗 CHAINED PROMISES VERSION
        async function chainPromises() {
            log('🎬 Starting Promise Chain...');

            try {
                // Chain 3 promises sequentially
                await delay(1000, '✅ MESSAGE 1: First promise resolved after 1s');
                await delay(2000, '✅ MESSAGE 2: Second promise resolved after 2s');
                await delay(3000, '✅ MESSAGE 3: Third promise resolved after 3s');

                log('🎉 All promises completed! Total time: ~6 seconds');
            } catch (error) {
                log(`❌ Error: ${error}`);
            }
        }

        // 📊 Alternative: Pure Promise Chain (.then)
        function chainWithThen() {
            log('🔄 Starting .then() chain...');

            delay(1000, '1️⃣ Then 1: 1s delay')
                .then(() => delay(2000, '2️⃣ Then 2: 2s delay'))
                .then(() => delay(3000, '3️⃣ Then 3: 3s delay'))
                .then(() => log('🎊 .then() chain complete!'))
                .catch(error => log(`❌ .then() Error: ${error}`));
        }

        // ⚡ Parallel vs Sequential comparison
        function compareParallel() {
            log('⚡ PARALLEL execution (all start at once):');

            const p1 = delay(1000, 'P1');
            const p2 = delay(2000, 'P2');
            const p3 = delay(3000, 'P3');

            // All complete after FASTEST (1s), not slowest (3s)
            Promise.all([p1, p2, p3]).then(() => {
                log('🏎️ Parallel complete! Total: 3s max');
            });
        }

        function log(message) {
            const now = new Date().toLocaleTimeString();
            const div = document.createElement('div');
            div.className = 'message';
            div.innerHTML = `[${now}] ${message}`;
            output.appendChild(div);
            output.scrollTop = output.scrollHeight;
            console.log(message);
        }

        function clearOutput() {
            output.innerHTML = '';
        }
    </script>
</body>
</html>

```

```python
State Transition Diagram

           ┌─────────────┐
           │   Pending   │  ← NEW Promise
           │             │
           └──────┬──────┘
                  │
       ┌───────────▼──────────┐
       │                      │
  ┌────▼────┐           ┌────▼────┐
  │Fulfilled│           │ Rejected│
  │(value)  │           │(reason) │
  └─────────┘           └─────────┘
     Terminal              Terminal

     <!DOCTYPE html>
<html>
<head>
    <title>Promise States Demo</title>
    <style>
        body { font-family: monospace; margin: 40px; line-height: 1.6; }
        .state { padding: 8px; margin: 4px 0; border-radius: 4px; font-weight: bold; }
        .pending { background: #ffc107; color: #856404; }
        .fulfilled { background: #28a745; color: white; }
        .rejected { background: #dc3545; color: white; }
        #output { background: #f8f9fa; padding: 20px; border-radius: 8px; min-height: 400px; }
        button { padding: 10px 20px; margin: 5px; cursor: pointer; }
    </style>
</head>
<body>
    <h1>🔮 Promise States & Transitions</h1>

    <div>
        <button onclick="demo1()">1️⃣ Basic States</button>
        <button onclick="demo2()">2️⃣ Multiple Scenarios</button>
        <button onclick="demo3()">3️⃣ State Inspector</button>
        <button onclick="clearOutput()">🗑️ Clear</button>
    </div>

    <h3>Rules:</h3>
    <ul>
        <li><code>Pending</code> → <code>Fulfilled</code> <strong>OR</strong> <code>Rejected</code></li>
        <li><strong>NO REVERSE</strong>: Fulfilled/Rejected are <em>terminal</em></li>
        <li><strong>ONLY ONCE</strong>: Can't change after settling</li>
    </ul>

    <div id="output"></div>

    <script>
        const output = document.getElementById('output');

        function log(message, stateClass = '') {
            const timestamp = new Date().toLocaleTimeString();
            const div = document.createElement('div');
            div.className = `state ${stateClass}`;
            div.textContent = `[${timestamp}] ${message}`;
            output.appendChild(div);
            output.scrollTop = output.scrollHeight;
            console.log(message);
        }

        // 1️⃣ Basic States Demo
        function demo1() {
            log('=== BASIC PROMISE STATES ===', '');
            output.innerHTML = '<h4>🔍 Live Trace:</h4>';

            const promise = new Promise((resolve, reject) => {
                log('🚀 Promise CREATED: Pending', 'pending');

                setTimeout(() => {
                    log('✅ resolve() called', 'fulfilled');
                    resolve('Success! 🎉');
                }, 1000);
            });

            // Check state progression
            promise
                .then(value => {
                    log(`✅ Fulfilled with: ${value}`, 'fulfilled');
                })
                .catch(error => {
                    log(`❌ Rejected with: ${error}`, 'rejected');
                });
        }

        // 2️⃣ Multiple Scenarios
        function demo2() {
            log('=== ALL TRANSITION SCENARIOS ===', '');

            // Scenario 1: Pending → Fulfilled
            new Promise(resolve => {
                log('📈 Pending → Fulfilled', 'pending');
                setTimeout(() => resolve('✅ Success'), 500);
            }).then(v => log(`Result: ${v}`, 'fulfilled'));

            // Scenario 2: Pending → Rejected
            new Promise((resolve, reject) => {
                log('📉 Pending → Rejected', 'pending');
                setTimeout(() => reject('❌ Error!'), 500);
            }).catch(e => log(`Error: ${e}`, 'rejected'));

            // Scenario 3: IMMEDIATE Fulfill
            log('⚡ Immediate Fulfill', 'fulfilled');
            Promise.resolve('Instant!').then(v => log(`Instant: ${v}`, 'fulfilled'));

            // Scenario 4: IMMEDIATE Reject
            log('💥 Immediate Reject', 'rejected');
            Promise.reject('Boom!').catch(e => log(`Boom: ${e}`, 'rejected'));
        }

        // 3️⃣ State Inspector (Advanced)
        function demo3() {
            log('=== IMMUTABLE STATE PROOF ===', '');

            const promise = new Promise(resolve => {
                log('1. Initial: Pending', 'pending');

                // Can't change after settling!
                setTimeout(() => {
                    resolve('Settled');
                    log('2. Now Fulfilled (Terminal)', 'fulfilled');

                    // These have NO EFFECT!
                    promise.resolve('Ignored!'); // Error!
                    promise.reject('Also ignored!');
                    log('3. Attempts to change state → IGNORED', 'fulfilled');
                }, 1000);
            });
        }

        function clearOutput() {
            output.innerHTML = '';
        }
    </script>
</body>
</html>
```

```python
<!DOCTYPE html>
<html>
<head>
    <title>Event Loop: Promises vs setTimeout</title>
    <style>
        body { font-family: monospace; margin: 40px; line-height: 1.6; }
        #timeline {
            background: #f8f9fa;
            padding: 20px;
            border-radius: 8px;
            font-size: 14px;
        }
        .microtask { background: #007bff; color: white; padding: 4px 8px; border-radius: 3px; }
        .macrotask { background: #28a745; color: white; padding: 4px 8px; border-radius: 3px; }
        .sync { background: #6c757d; color: white; padding: 4px 8px; border-radius: 3px; }
        button { padding: 12px 24px; font-size: 16px; cursor: pointer; margin: 10px 0; }
    </style>
</head>
<body>
    <h1>⚙️ Event Loop: Microtask vs Macrotask Priority</h1>

    <button onclick="runDemo()">▶️ Run Event Loop Demo</button>
    <button onclick="clearOutput()">🗑️ Clear</button>

    <h3>Prediction:</h3>
    <pre>
1. Sync code
2. ALL Promises (.then)  ← Microtask Queue
3. setTimeout callbacks  ← Macrotask Queue
    </pre>

    <div id="output"></div>

    <div id="timeline"></div>

    <script>
        const output = document.getElementById('output');
        const timeline = document.getElementById('timeline');

        function log(type, message) {
            const timestamp = Date.now();
            const div = document.createElement('div');
            div.innerHTML = `<span class="${type}">[${timestamp}]</span> ${message}`;
            output.appendChild(div);
            timeline.innerHTML += `<span class="${type}">${type[0].toUpperCase()}</span> `;
            output.scrollTop = output.scrollHeight;
            console.log(`[${timestamp}] ${type}: ${message}`);
        }

        function runDemo() {
            output.innerHTML = '<h4>🔍 Execution Order:</h4>';
            timeline.innerHTML = '';

            log('sync', '1️⃣ SYNCHRONOUS CODE executes first');

            // Microtasks scheduled
            Promise.resolve().then(() => {
                log('microtask', '2️⃣ Promise .then() (Microtask #1)');
                return Promise.resolve();
            }).then(() => {
                log('microtask', '3️⃣ Promise .then() (Microtask #2)');
            });

            // More microtasks
            queueMicrotask(() => {
                log('microtask', '4️⃣ queueMicrotask()');
            });

            // Macrotasks scheduled
            setTimeout(() => {
                log('macrotask', '5️⃣ setTimeout(0) (Macrotask #1)');
            }, 0);

            setTimeout(() => {
                log('macrotask', '6️⃣ setTimeout(0) (Macrotask #2)');
            }, 0);

            // MutationObserver (also microtask)
            const observer = new MutationObserver(() => {
                log('microtask', '7️⃣ MutationObserver (Microtask)');
            });
            observer.observe(document.body, { attributes: true });
            document.body.setAttribute('data-demo', 'observed');

            log('sync', '🏁 Sync code complete - Event Loop processes queues');
        }

        function clearOutput() {
            output.innerHTML = '';
            timeline.innerHTML = '';
        }
    </script>
</body>
</html>

                event loop phase
                Event Loop Cycle:
1. ┌─ Execute Sync ───────┐
   │ console.log("A")    │
   └─────────────────────┘
2. ┌─ Microtasks ────────┐  ← Promises, queueMicrotask, MutationObserver
   │ .then()             │  ⭐⭐⭐⭐⭐ HIGHEST PRIORITY
   └─────────────────────┘
3. ┌─ Render (if needed)─┐
   │ Update UI           │
   └─────────────────────┘
4. ┌─ Macrotasks ────────┐  ← setTimeout, setInterval, I/O, UI events
   │ setTimeout(0)       │  ⭐⭐⭐ Lower Priority
   └─────────────────────┘
5. Repeat ⏭️
```
