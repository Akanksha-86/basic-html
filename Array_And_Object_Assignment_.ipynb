````python
Sure! Let’s use a **`for...in` loop** to iterate over the `student` object and perform some useful operations, like printing each property and its value. I’ll also show a few examples of how you can manipulate or check the values.

---

## ✅ Example: Iterating with `for...in`

```javascript id="u9f1mn"
const student = {
  name: "Alice",
  age: 22,
  major: "Computer Science",
  GPA: 3.8,
  isEnrolled: true
};

// 1. Print all properties and values
for (let key in student) {
  console.log(`${key}: ${student[key]}`);
}

// 2. Check if student is enrolled
for (let key in student) {
  if (key === "isEnrolled" && student[key] === true) {
    console.log("The student is currently enrolled.");
  }
}

// 3. Calculate if GPA is above 3.5
for (let key in student) {
  if (key === "GPA" && student[key] > 3.5) {
    console.log("The student has a high GPA.");
  }
}

// 4. List only string type properties
for (let key in student) {
  if (typeof student[key] === "string") {
    console.log(`String property - ${key}: ${student[key]}`);
  }
}
```

---

## 🔍 Explanation

1. **`for...in`** iterates over the **keys of the object** (`name`, `age`, etc.).
2. You access the **value** with `student[key]`.
3. You can perform **conditions or operations** inside the loop, like checking `GPA` or `isEnrolled`.
4. You can also filter properties based on **type**, value, or any other logic.

---

## 💡 Output of Example:

```
name: Alice
age: 22
major: Computer Science
GPA: 3.8
isEnrolled: true
The student is currently enrolled.
The student has a high GPA.
String property - name: Alice
String property - major: Computer Science

- (a) Here’s a clean JavaScript function that does exactly what you described using a **`for...in` loop**:

---

```javascript id="q1student"
const student = {
  name: "Alice",
  age: 22,
  major: "Computer Science",
  GPA: 3.8,
  isEnrolled: true
};

function displayStudentInfo(studentObj) {
  for (let key in studentObj) {
    console.log(`Property: ${key}, Value: ${studentObj[key]}`);
  }
}

// Call the function
displayStudentInfo(student);
```

---

## 🔍 Explanation

1. `studentObj` is the parameter for the function, so it can work with **any student object**.
2. `for (let key in studentObj)` iterates over all property names.
3. `studentObj[key]` gives the value corresponding to that property.
4. `console.log` prints each property-value pair in the requested format.

---

## 💡 Output

```
Property: name, Value: Alice
Property: age, Value: 22
Property: major, Value: Computer Science
Property: GPA, Value: 3.8
Property: isEnrolled, Value: true
````
