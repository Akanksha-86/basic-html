2. Explain the role of the following key properties in the flexbox layout model
 - 1. **Justify-content:** the justify-content property is used to align flex items or grid items along the main axis of their container. it controls how extra space within the container is distributed, either between items or around them, when the items don't fill the entire space.

  2. **Align-items:** align-item is a property that control the alignment of flex item along the cross axis(perpendicular to the main axis ) within a flex container.
  3. **gap:** the gap property define the size of the gap between the row and column in flexbox, grid or multi column layout. it is a shorthand fofr the following properties: row-gap, column-gap.
  4. **flex-direction:**the flex-direction CSS property set how flex items are placed in the flex container defining the main axis and the direction (normal or reversed).
  5. **flex-wrap:**the flex-direction CSS property sets weather fixed items are forced onto one line can wrap onto multiple lines. if wrapping is allowed, it setsthe direction that lines are stacked.
