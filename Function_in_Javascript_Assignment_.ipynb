```python
<!DOCTYPE html>
<html>
<head>
    <title>Recursive Factorial Calculator</title>
    <style>
        * { box-sizing: border-box; margin: 0; padding: 0; }
        body {
            font-family: 'Segoe UI', system-ui, sans-serif;
            min-height: 100vh;
            background: linear-gradient(135deg, #1e1b4b 0%, #2a1d5c 50%, #0f0f29 100%);
            display: flex;
            flex-direction: column;
            align-items: center;
            justify-content: center;
            padding: 20px;
            color: white;
        }
        .container {
            background: rgba(255,255,255,0.08);
            backdrop-filter: blur(25px);
            border-radius: 24px;
            padding: 40px;
            max-width: 600px;
            width: 100%;
            box-shadow: 0 30px 60px rgba(0,0,0,0.5);
            border: 1px solid rgba(255,255,255,0.15);
            text-align: center;
        }
        h1 {
            font-size: 2.8em;
            font-weight: 900;
            background: linear-gradient(45deg, #00ff88, #00ccff, #ff6b9d);
            -webkit-background-clip: text;
            background-clip: text;
            margin-bottom: 15px;
            background-size: 200% 200%;
            animation: gradientShift 3s ease infinite;
        }
        @keyframes gradientShift {
            0%, 100% { background-position: 0% 50%; }
            50% { background-position: 100% 50%; }
        }
        .input-group {
            display: flex;
            gap: 15px;
            margin: 30px 0;
            flex-wrap: wrap;
            justify-content: center;
        }
        input {
            padding: 18px 24px;
            font-size: 20px;
            border: none;
            border-radius: 16px;
            background: rgba(255,255,255,0.2);
            color: white;
            text-align: center;
            min-width: 120px;
            backdrop-filter: blur(10px);
        }
        input::placeholder { color: rgba(255,255,255,0.6); }
        button {
            padding: 18px 32px;
            font-size: 18px;
            font-weight: 700;
            background: linear-gradient(45deg, #ff6b6b, #ffd93d);
            color: #1e1b4b;
            border: none;
            border-radius: 16px;
            cursor: pointer;
            transition: all 0.3s;
            box-shadow: 0 10px 30px rgba(255,107,107,0.4);
        }
        button:hover {
            transform: translateY(-4px) scale(1.02);
            box-shadow: 0 20px 45px rgba(255,107,107,0.6);
        }
        .result {
            background: linear-gradient(135deg, rgba(0,255,136,0.2), rgba(0,204,255,0.2));
            border: 3px solid rgba(0,255,136,0.6);
            padding: 35px;
            border-radius: 24px;
            margin: 30px 0;
            font-size: 32px;
            font-weight: 900;
            text-shadow: 0 2px 10px rgba(0,0,0,0.5);
        }
        .recursion-tree {
            background: rgba(0,0,0,0.3);
            border-radius: 16px;
            padding: 25px;
            margin: 30px 0;
            font-family: 'Courier New', monospace;
            font-size: 14px;
            text-align: left;
            max-height: 200px;
            overflow-y: auto;
            border-left: 5px solid #00ff88;
        }
        .test-buttons {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(140px, 1fr));
            gap: 12px;
            margin-top: 30px;
        }
        .test-btn {
            padding: 15px;
            background: rgba(255,255,255,0.15);
            border: 2px solid rgba(255,255,255,0.3);
            border-radius: 12px;
            color: white;
            cursor: pointer;
            font-weight: 600;
            transition: all 0.3s;
            font-size: 16px;
        }
        .test-btn:hover {
            background: rgba(255,255,255,0.25);
            transform: translateY(-2px);
            box-shadow: 0 10px 25px rgba(0,0,0,0.4);
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>🔢 Recursive Factorial</h1>
        <p style="opacity: 0.9; margin-bottom: 30px; font-size: 1.2em;">
            n! = n × (n-1) × (n-2) × ... × 1
        </p>

        <div class="input-group">
            <input type="number" id="numInput" placeholder="Enter n (0-20)" min="0" max="20" value="5">
            <button onclick="calculateFactorial()">Calculate n! ➕</button>
        </div>

        <!-- Result -->
        <div id="result" class="result" style="display: none;">
            <span id="resultText">120</span>
        </div>

        <!-- Recursion Visualization -->
        <div id="recursionTree" class="recursion-tree" style="display: none;"></div>

        <!-- Test Buttons -->
        <div class="test-buttons">
            <button class="test-btn" onclick="testFactorial(0)">0! = 1</button>
            <button class="test-btn" onclick="testFactorial(1)">1! = 1</button>
            <button class="test-btn" onclick="testFactorial(5)">5! = 120</button>
            <button class="test-btn" onclick="testFactorial(10)">10! = 3.6M</button>
        </div>
    </div>

    <script>
    // 🏹 RECURSIVE FACTORIAL FUNCTION
    function factorial(n) {
        // Base cases
        if (n < 0) return "❌ Error: Negative numbers not allowed";
        if (n === 0 || n === 1) return 1;

        // Recursive case: n! = n × (n-1)!
        return n * factorial(n - 1);
    }

    // 🎯 UI Calculation
    function calculateFactorial() {
        const n = parseInt(document.getElementById('numInput').value);
        const result = factorial(n);

        document.getElementById('resultText').textContent = result;
        document.getElementById('result').style.display = 'block';

        // Show recursion tree
        showRecursionTree(n);

        console.log(`factorial(${n}) = ${result}`);
    }

    function testFactorial(n) {
        document.getElementById('numInput').value = n;
        calculateFactorial();
    }

    function showRecursionTree(n) {
        let tree = `📊 Recursion Tree for ${n}!\n`;
        tree += "─".repeat(40) + "\n";

        function buildTree(current, depth = 0) {
            const indent = "  ".repeat(depth);
            if (current <= 1) {
                tree += `${indent}📍 ${current}! = 1\n`;
                return 1;
            }
            tree += `${indent}📌 ${current}! = ${current} × ${current-1}!\n`;
            return current * buildTree(current - 1, depth + 1);
        }

        buildTree(n);
        document.getElementById('recursionTree').textContent = tree;
        document.getElementById('recursionTree').style.display = 'block';
    }

    // 🧪 TEST WITH DIFFERENT INPUTS
    console.log("🔢 Factorial Tests:");
    console.log("factorial(0) =", factorial(0));   // 1
    console.log("factorial(1) =", factorial(1));   // 1
    console.log("factorial(5) =", factorial(5));   // 120
    console.log("factorial(10) =", factorial(10)); // 3628800

    // Initial demo
    calculateFactorial();
    </script>
</body>
</html>
```

```python
<!DOCTYPE html>
<html>
<head>
    <title>Curry Function Demo</title>
    <style>
        * { box-sizing: border-box; margin: 0; padding: 0; }
        body {
            font-family: 'Inter', -apple-system, BlinkMacSystemFont, sans-serif;
            min-height: 100vh;
            background: linear-gradient(135deg, #ff6b6b 0%, #feca57 50%, #ff9ff3 100%);
            padding: 20px;
            color: #2d3436;
        }
        .container {
            max-width: 900px;
            margin: 0 auto;
            background: rgba(255,255,255,0.95);
            backdrop-filter: blur(20px);
            border-radius: 24px;
            padding: 40px;
            box-shadow: 0 30px 80px rgba(255,107,107,0.4);
            border: 1px solid rgba(255,255,255,0.8);
        }
        h1 {
            text-align: center;
            font-size: 3em;
            font-weight: 900;
            background: linear-gradient(45deg, #2d3436, #636e72);
            -webkit-background-clip: text;
            background-clip: text;
            margin-bottom: 15px;
        }
        .curry-demo {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 40px;
            margin: 40px 0;
        }
        .demo-section {
            background: white;
            padding: 30px;
            border-radius: 20px;
            box-shadow: 0 15px 40px rgba(0,0,0,0.1);
            border: 3px solid rgba(255,255,255,0.5);
        }
        .demo-section h3 {
            font-size: 1.5em;
            margin-bottom: 20px;
            color: #2d3436;
        }
        input {
            width: 100%;
            padding: 16px 20px;
            font-size: 18px;
            border: 2px solid #dfe6e9;
            border-radius: 12px;
            margin: 10px 0;
            transition: all 0.3s;
        }
        input:focus {
            outline: none;
            border-color: #00b894;
            box-shadow: 0 0 0 4px rgba(0,184,148,0.1);
        }
        button {
            width: 100%;
            padding: 16px;
            font-size: 16px;
            font-weight: 700;
            background: linear-gradient(45deg, #00b894, #00a085);
            color: white;
            border: none;
            border-radius: 12px;
            cursor: pointer;
            transition: all 0.3s;
            margin: 10px 0;
        }
        button:hover {
            transform: translateY(-2px);
            box-shadow: 0 10px 30px rgba(0,184,148,0.4);
        }
        .result {
            background: linear-gradient(135deg, #00b894, #00a085);
            color: white;
            padding: 25px;
            border-radius: 16px;
            font-size: 24px;
            font-weight: 800;
            text-align: center;
            margin-top: 20px;
            box-shadow: 0 10px 30px rgba(0,184,148,0.3);
        }
        .curry-steps {
            background: #f8f9fa;
            padding: 20px;
            border-radius: 12px;
            margin-top: 20px;
            font-family: 'Courier New', monospace;
            font-size: 14px;
            text-align: left;
            border-left: 5px solid #00b894;
        }
        .code-block {
            background: #2d3436;
            color: #00ff88;
            padding: 25px;
            border-radius: 16px;
            font-family: 'Courier New', monospace;
            font-size: 15px;
            line-height: 1.7;
            margin: 30px 0;
            overflow-x: auto;
        }
        .test-cases {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 15px;
            margin-top: 30px;
        }
        .test-btn {
            padding: 15px;
            background: rgba(255,255,255,0.7);
            border: 2px solid #dfe6e9;
            border-radius: 12px;
            cursor: pointer;
            font-weight: 600;
            transition: all 0.3s;
        }
        .test-btn:hover {
            background: white;
            border-color: #00b894;
            transform: translateY(-2px);
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>🌶️ Curry Function</h1>
        <p style="text-align: center; font-size: 1.3em; opacity: 0.9; margin-bottom: 40px;">
            Transform add(a, b) → add(a)(b) using closures
        </p>

        <!-- CURRY CODE -->
        <div class="code-block">
            <strong>🏹 Curry Function Implementation:</strong><br><br>
            <code>function curry(func) {<br>
            &nbsp;&nbsp;return function curried(...args) {<br>
            &nbsp;&nbsp;&nbsp;&nbsp;if (args.length >= func.length) {<br>
            &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return func(...args);<br>
            &nbsp;&nbsp;&nbsp;&nbsp;}<br>
            &nbsp;&nbsp;&nbsp;&nbsp;return function(...nextArgs) {<br>
            &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return curried(...args, ...nextArgs);<br>
            &nbsp;&nbsp;&nbsp;&nbsp;};<br>
            &nbsp;&nbsp;};<br>
            }</code>
        </div>

        <!-- Demo Sections -->
        <div class="curry-demo">
            <!-- Normal Add -->
            <div class="demo-section">
                <h3>📊 Normal Function</h3>
                <div style="display: flex; gap: 10px;">
                    <input type="number" id="num1" placeholder="a" value="5">
                    <input type="number" id="num2" placeholder="b" value="3">
                </div>
                <button onclick="normalAdd()">add(5, 3) ➕</button>
                <div id="normalResult" class="result" style="display:none;"></div>
            </div>

            <!-- Curried Add -->
            <div class="demo-section">
                <h3>🌶️ Curried Function</h3>
                <input type="number" id="curryNum1" placeholder="a" value="5">
                <button onclick="curryStep1()">add(a) →</button>
                <input type="number" id="curryNum2" placeholder="b" value="3">
                <button onclick="curryStep2()">➕ add(a)(b)</button>
                <div id="curryResult" class="result" style="display:none;"></div>
                <div id="currySteps" class="curry-steps" style="display:none;"></div>
            </div>
        </div>

        <!-- Test Cases -->
        <div class="test-cases">
            <button class="test-btn" onclick="testCurry(3,4)">3 + 4 = 7</button>
            <button class="test-btn" onclick="testCurry(10,15)">10 + 15 = 25</button>
            <button class="test-btn" onclick="testCurry(20,30)">20 + 30 = 50</button>
            <button class="test-btn" onclick="testCurry(7,8)">7 + 8 = 15</button>
        </div>
    </div>

    <script>
    // 🏹 CURRY FUNCTION IMPLEMENTATION
    function curry(func) {
        return function curried(...args) {
            // If all arguments received, execute original function
            if (args.length >= func.length) {
                return func(...args);
            }
            // Return new function expecting more arguments
            return function(...nextArgs) {
                return curried(...args, ...nextArgs);
            };
        };
    }

    // ➕ Original add function
    function add(a, b) {
        return a + b;
    }

    // 🌶️ Curried version
    const curriedAdd = curry(add);

    // 📊 Normal add
    function normalAdd() {
        const a = parseFloat(document.getElementById('num1').value) || 0;
        const b = parseFloat(document.getElementById('num2').value) || 0;
        const result = add(a, b);

        document.getElementById('normalResult').innerHTML = `add(${a}, ${b}) ➕ <strong>${result}</strong>`;
        document.getElementById('normalResult').style.display = 'block';
    }

    // 🌶️ Curry step 1
    let partialAdd;
    function curryStep1() {
        const a = parseFloat(document.getElementById('curryNum1').value) || 0;
        partialAdd = curriedAdd(a); // Returns function waiting for b

        document.getElementById('currySteps').innerHTML = `
            <strong>Step 1:</strong> curriedAdd(${a})<br>
            <code>➜ partial function: add(${a}, ?)</code>
        `;
        document.getElementById('currySteps').style.display = 'block';
        document.getElementById('curryResult').style.display = 'none';
    }

    // 🌶️ Curry step 2
    function curryStep2() {
        if (!partialAdd) {
            alert('First click "add(a) →"');
            return;
        }

        const b = parseFloat(document.getElementById('curryNum2').value) || 0;
        const result = partialAdd(b); // Executes original add(a, b)

        document.getElementById('curryResult').innerHTML = `
            add(${partialAdd.a || ''}, ${b}) ➕ <strong>${result}</strong>
        `;
        document.getElementById('curryResult').style.display = 'block';

        document.getElementById('currySteps').innerHTML += `
            <br><strong>Step 2:</strong> partialAdd(${b})<br>
            <code>➜ add(${partialAdd.a || ''}, ${b}) = ${result}</code>
        `;
    }

    function testCurry(a, b) {
        document.getElementById('num1').value = a;
        document.getElementById('num2').value = b;
        document.getElementById('curryNum1').value = a;
        document.getElementById('curryNum2').value = b;

        normalAdd();
        curryStep1();
        setTimeout(curryStep2, 500);
    }

    // 🧪 CONSOLE TESTS
    console.log("🌶️ Curry Tests:");
    console.log("Normal: add(5, 3) =", add(5, 3));
    console.log("Curried: curriedAdd(5)(3) =", curriedAdd(5)(3));
    console.log("Chained: curriedAdd(2)(3)(4) =", curriedAdd(2)(3)(4)); // Works with extra args!

    // Initial demo
    testCurry(5, 3);
    </script>
</body>
</html>
```
